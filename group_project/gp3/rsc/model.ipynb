```python
user_tweets = read_from_csv('user_tweets_only_numeric_feature_engineer.csv', PROCESSED_DATA_DIR)
```

```python
user_tweets
```

```python
user_tweets = shuffle(user_tweets, random_state=0)
```

```python
y = user_tweets['state_back']
X = user_tweets.drop(columns=['state_back'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7)
```

```python
X_train
```

```python
clf = RandomForestClassifier(n_estimators=25)
clf.fit(X_train, y_train)
clf_probs = clf.predict_proba(X_test)
```

```python
pred = clf.predict(X_test)
pred_train = clf.predict(X_train)
```

```python
def matrix(y_test, pred):
    print('F1 Score on Test Set: {}'.format(metrics.f1_score(np.array(y_test), pred)))
    print('F1 Score on Training Set: {}'.format(metrics.f1_score(np.array(y_train), pred_train)))
    print('Recall Score on Test Set: {}'.format(metrics.recall_score(np.array(y_test), pred)))
    print('Recall Score on Training Set: {}'.format(metrics.recall_score(np.array(y_train), pred_train)))
    print('Precision Score on Test Set: {}'.format(metrics.precision_score(np.array(y_test), pred)))
    print('Precision Score on Training Set: {}'.format(metrics.precision_score(np.array(y_train), pred_train)))
matrix(y_test, pred)
```
